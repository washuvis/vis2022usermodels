```python
commands = np.array([])

for dataset in user_interaction_data_paths:
    interaction_data = None
    if dataset == 'stl':
        interaction_data = pd.read_csv(user_interaction_data_paths[dataset])
        
    elif dataset == 'vast':
        interaction_data = pd.read_pickle(user_interaction_data_paths[dataset])
        interaction_data = interaction_data[interaction_data.experimental_group=='control'].copy()
        interaction_data = interaction_data.reset_index(drop=True)

    elif dataset == 'boardrooms':
        interaction_data = pd.read_csv(user_interaction_data_paths[dataset])

    elif dataset == 'political':
        interaction_data = pd.read_csv(user_interaction_data_paths[dataset])
        
    for session in interaction_data.index:
        string = f'python3 ensemble_run.py {dataset} {session}'
        commands = np.append(commands, string)
```

```python
np.random.shuffle(commands)
```

```python
for s in commands:
    print('[] ', s)
```
